```python
data = []

CX = np.array(imgC).reshape(np.array(imgC).shape[0],-1)
CY = [2]*len(imgC)
CY = np.array(CY).reshape(-1,1)


DX = np.array(imgD).reshape(np.array(imgD).shape[0],-1)
DY = [1]*len(imgD)
DY = np.array(DY).reshape(-1,1)

for i in range(len(CY)):
    data.append([CX[i], CY[i]])

for i in range(len(DY)):
    data.append([DX[i], DY[i]])

```

```python
train = data[:16100]
test = data[16100:]

train_data =[]
for i in range(len(train)):
    train_data.append(train[i][0])

train_label = []
for i in range(len(train)):
    train_label.append(train[i][1])
    
print(len(test))
print(len(train))
print(len(data))

test_data =[]
for i in range(len(test)):
    test_data.append(train[i][0])


test_label = []
for i in range(len(test)):
    test_label.append(train[i][1])
    
# for i in range(len(test)):
#     print(test_label[i])
# type(train[1][1])
```

```python
def next_batch(batch_size):
    _index_in_epoch = 0
    start = _index_in_epoch
    _index_in_epoch += batch_size
    _epochs_done = 0

    if _index_in_epoch > len(train):
        _epochs_done += 1
        start = 0
        _index_in_epoch = batch_size
    assert batch_size <= len(train)
    end = _index_in_epoch
    return train_data[start:end], train_label[start:end]
```

```python
a, b  = next_batch(20)

# train_label[:20]
```

```python
import tensorflow as tf
```

```python
x = tf.placeholder(tf.float32, [None, 4096])
```

```python
#model parameters
#W and b initalized as a tensor full of zeros
W = tf.Variable(tf.zeros([4096, 1]))
b = tf.Variable(tf.zeros([1]))
```

```python
#implement our model
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
#To implement cross-entropy we need to first add a new placeholder to input the correct answers
#cross-entropy is loss or cost function
y_ = tf.placeholder(tf.float32, [None, 1])
```

```python
#implement cross entropy function
#First, tf.log computes the logarithm of each element of y
#Next, we multiply each element of y_ with the corresponding element of tf.log(y)
#Then tf.reduce_sum adds the elements in the second dimension of y, due to the reduction_indices=[1] parameter.
#Finally, tf.reduce_mean computes the mean over all the examples in the batch.
#cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))

cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

```python
#back propagation and gradient descent
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
#launch model in interactive session
sess = tf.InteractiveSession()
```

```python
#create an operation to initialize the variables we created
tf.global_variables_initializer().run()
```

```python
#run the training step 1000 times
for _ in range(100):
    batch_xs, batch_ys = next_batch(10)
    sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```
